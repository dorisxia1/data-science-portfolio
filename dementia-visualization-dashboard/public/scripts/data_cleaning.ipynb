```python
# import libraries
import pandas as pd
```

```python
# load the file
data = pd.read_csv('../public/data/raw/NORC_Dementia_DataHub_Nat_State_2020.csv')
data.head(3)
```

```python
# the other categories not used in the plot need to be set to null in order to gain accurate count

df = data[(data['FIPS_STATE'].notna()) & (data['INSURANCE'].notna()) & (data['RACECAT'].notna()) & \
          (data['prev_cnt'].notna()) & (data['mort_cnt'].notna()) & (data['AGECAT2'].isna()) & (data['AGECAT6'].isna()) & \
          (data['SEXCAT'].isna()) & (data['DEMENTIACAT'].isna())]\
          [['FIPS_STATE', 'INSURANCE', 'RACECAT', 'prev_cnt', 'mort_cnt']].copy(deep=True)
```

```python
import csv
```

```python
input_file = '../public/data/raw/state-code-book.txt'
```

```python
output_file = '../public/data/cleaned/fips_states_codebook.csv'
```

```python
rows = []
with open(input_file, 'r') as f:
    for line in f:
        code, state = line.split('=')
        rows.append([state.strip(), code.strip()])
```

```python
with open(output_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['State', 'Code'])  # write headers
    writer.writerows(rows)
```
