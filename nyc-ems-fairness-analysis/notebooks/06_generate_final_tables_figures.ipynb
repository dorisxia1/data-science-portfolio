# Generating Final Tables and Figures

```python
import os
import pandas as pd
import glob
import shutil
```

### Move tables to ../outputs/tables/raw

```python
raw_dir = "../outputs/tables/raw"
final_dir = "../outputs/tables/final"

os.makedirs(raw_dir, exist_ok=True)
os.makedirs(final_dir, exist_ok=True)
```

```python
base_dir = "../outputs/tables"

files_to_move = glob.glob(os.path.join(base_dir, "*.csv"))
if not files_to_move:
    print("No CSV files found in the base directory to move.")
else:
    for file_path in files_to_move:
        file_name = os.path.basename(file_path)
        dest_path = os.path.join(raw_dir, file_name)
        
        # Move the file
        shutil.move(file_path, dest_path)
        print(f"Moved: {file_name} -> outputs/tables/raw/")
```

Conditional Summary by Tier High Black Zip

```python
cond_summ_high_black = pd.read_csv("../outputs/tables/raw/conditional_summary_by_tier_high_black_zip.csv")
cond_summ_high_black
```

```python
cond_summ_high_black.columns = ["Severity Tier", "High Black Population ZIP (Yes = 1)", "n", "Mean", "Median", "Std. Dev"]
severity_order = ['High', 'Medium', 'Low']
cond_summ_high_black['Severity Tier'] = pd.Categorical(cond_summ_high_black['Severity Tier'], categories=severity_order, ordered=True)
cond_summ_high_black_sorted = cond_summ_high_black.sort_values(by=['Severity Tier', 'High Black Population ZIP (Yes = 1)'])
cond_summ_high_black_sorted
```
